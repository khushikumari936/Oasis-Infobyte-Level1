```python
data1.columns=data1.columns.str.strip()
data1.columns=data1.columns.str.replace(' ', '_')
data1.columns
```

```python
data1['Calorie_Level'] = pd.cut(
    data1['Calories'],
    bins=[0, 300, 600, 1000],
    labels=['Low', 'Medium', 'High']
)
```

```python
data1['Healthy_Item'] = np.where(
    (data1['Total_Fat'] < 15) &
    (data1['Sugars'] < 10),
    "Yes",
    "No"
)
```
